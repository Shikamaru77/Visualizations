## Power BI 
A suite of business analytics tools that allows users to analyze data and share insights. It is a cloud-based service that can be accessed from anywhere and on any device, and it provides a wide range of features and functionalities to help users visualize and analyze data, and share insights with others.

### Power Query

Power Query is a data transformation and analysis tool that is integrated into Power BI, Microsoft's business intelligence platform. In Power BI, Power Query is known as the "Query Editor", and it allows users to easily connect to and transform data from a wide variety of sources, including databases, web pages, text files, and Excel workbooks.

### Power Pivot


Power Pivot is a data modeling and analysis tool that is integrated into Power BI, Microsoft's business intelligence platform. It allows users to analyze and visualize large amounts of data from multiple sources, and create powerful data models that can be used to build reports and dashboards.
